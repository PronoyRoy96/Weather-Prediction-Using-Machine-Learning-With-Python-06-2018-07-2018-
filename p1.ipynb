```python
feat_list=['maxtempm','mintempm','meantempm_1','maxtempm_1','mintempm_1','mindewptm_1','meandewptm_1','maxdewptm_1','meantempm_2',
'maxtempm_2',
'meantempm_3',
'mintempm_2',
'mindewptm_2',
'meandewptm_2',
'maxdewptm_2',
'mintempm_3',
'meandewptm_3',
'mindewptm_3',
'maxtempm_3',
'maxdewptm_3']
feat_selected=pd.DataFrame()

```

```python
import sklearn 
from sklearn import linear_model as lm
from sklearn import model_selection as ms
from sklearn import metrics as met
```

```python
train_feat,test_feat,train_target,test_target=ms.train_test_split(feat_selected,target,test_size =0.25)

print(len(train_feat),len(train_target))
print(len(test_feat),len(test_target))
```

```python
predictor=lm.LinearRegression()
```

```python
predictor.fit(train_feat,train_target)
```

```python
pred_y=predictor.predict(test_feat)
```

```python
score=met.r2_score(test_target,pred_y)
score
```

```python
mse=met.mean_squared_error(test_target,pred_y)
mse
```

```python
mae=met.mean_absolute_error(test_target,pred_y)
mae
```
